```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel("Investments_analysis.xlsx")
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_data=df.iloc[:,:]
df_data
sns.pairplot(df_data,diag_kind='kde')
plt.savefig("pair_plots.png",dpi=1200)
```

```python
correlation=df_data.corr()
sns.heatmap(correlation,annot=True)
plt.savefig("heatmap_correlation.png",dpi=1200)
plt.show()
```

```python
df2=pd.DataFrame(correlation)
df2.to_excel("Correlation.xlsx")
```
